```python
import cv2
import os
import numpy as np
```

```python
path_normal='./chest_xray/train/normal'
```

```python
images_normal=os.listdir(path_normal)
```

```python
images_normal
```

```python
cnt=1
labels=list()
images=list()
for image in images_normal:
    print(str(cnt)+'/1341')
    im=cv2.imread(path_normal+'/'+image)
    im=cv2.resize(im,(300,300),interpolation = cv2.INTER_NEAREST)
    images.append(im)
    labels.append(0)
    cnt+=1
```

```python
path_affected='./chest_xray/train/pneumonia'
```

```python
affected_images=os.listdir(path_affected)
```

```python
len(affected_images)
```

```python
cnt=1
for image in affected_images:
    print(str(cnt)+'/3875')
    im=cv2.imread(path_affected+'/'+image)
    im=cv2.resize(im,(300,300),interpolation = cv2.INTER_NEAREST)
    images.append(im)
    labels.append(1)
    cnt+=1
```

```python
images=np.array(images)
labels=np.array(labels)
```

```python
np.save('npy_data2/images.npy',np.array(images))
np.save('npy_data2/labels.npy',np.array(labels))
```

```python
path_test_normal='./chest_xray/test/normal'
path_test_affected='./chest_xray/test/pneumonia'
path_validation_normal='./chest_xray/val/normal'
path_validation_affected='./chest_xray/val/pneumonia'
```

```python
test_images_normal=os.listdir(path_test_normal)
test_images_affected=os.listdir(path_test_affected)
```

```python
test_images=list()
test_labels=list()
cnt=1
for i in test_images_normal:
    im=cv2.imread(path_test_normal+'/'+i)
    im=cv2.resize(im,(300,300),interpolation = cv2.INTER_NEAREST)
    test_images.append(im)
    print(str(cnt)+'/'+str(len(test_images_normal)))
    test_labels.append(0)
    cnt+=1
```

```python
cnt=1
for i in test_images_affected:
    im=cv2.imread(path_test_affected+'/'+i)
    im=cv2.resize(im,(300,300),interpolation = cv2.INTER_NEAREST)
    test_images.append(im)
    print(str(cnt)+'/'+str(len(test_images_affected)))
    test_labels.append(1)
    cnt+=1
```

```python
test_images=np.array(test_images)
test_labels=np.array(test_labels)
```
